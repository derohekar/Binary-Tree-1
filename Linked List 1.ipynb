```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
```

```python
def printll(head):
    
    while head is not None:
        print(str(head.data) + "->", end=" ")
        head = head.next
    print("None")
    return
```

```python
def takeinput():
    
    inputlist = [int(ele) for ele in input().split()]
    head = None
    for currdata in inputlist :
        if currdata == -1:
            break
            
        newNode = Node(currdata)
        if head is None:
                 head = newNode
   
        
        else:
            curr = head
            while curr.next is not None:
                curr = curr.next
            curr.next = newNode
            
    return head
```

```python
# insert at Ith position----> iteratively

def length(head):
    count = 0
    while head is not None:
        count += 1
        head = head.next
    return count

def insertIteratively(head, i, data):
    
    if i < 0 or i > length(head):
        return head
    count = 0
    prev = None
    curr = head
    
    while count < i :
        prev = curr
        curr = curr.next
        count += 1
        
    newNode = Node(data)
    while prev is not None:
        prev.next = newNode
        
    else:
        head = newNode
    newNode.next = curr
    return head
```

```python
# insert at Ith position----> Recursively

def insertAtIthRecursive(head, pos, data):
    if pos < 0:
        return head
    if pos == 0:
        newNode = Node(data)
        newNode.next = head
        return newNode
# pos == 0 step made before head == None, beacause if head is None and i == 0,  it will create a newNode
    
    if head is None:
        return None
    smallhead = insertAtIthRecursive(head.next, pos - 1, data)
    head.next = smallhead
    return head
    
```

```python
# delete a node Iteratively

def deleteNode(head, pos) :
    if head == None:
        return 
    
    if pos < 0 or pos >=length(head):
        return head

    count = 0
    prev = None
    curr = head

    while count < pos:
        prev = curr
        curr = curr.next
        count += 1
    if count == 0:
        head = head.next
        return head

    prev.next = curr.next
    curr.next = None
    return head
```

```python
# delete a node Iteratively

def deleteNodeRecursively(head, pos):
    if pos == 0:
        return head.next
    if head.next is None:
        return head
    smallhead = deleteNodeRecursively(head.next, pos - 1) 
    head.next = smallhead
    return head
```

```python
# Find a node in Linked List ----> iterative

def findNode(head, n):
    count = 0
    while head is not None:
        
        if head.data == n:
            return count
        count += 1
        head = head.next
    return -1
    
```

```python
# Find a node in Linked List ----> Recursive

def findNodeRecursive(head, n):
    if head == None:
        return -1
    if head.data == n:
        return 0
    smallhead = findNodeRecursive(head.next, n)
    if smallhead != -1:
        return 1 + smallhead
    return -1
```

```python
# Remove Duplicates

def removeDuplicates(head):
 
    # do nothing if the list is empty
    if head is None or head.next is None:
        return None
 
    current = head
    prev = None
 
    # compare the current node with the next node
    while current.next is not None:
        if current.data == current.next.data:
            prev = current.next.next
            current.next = prev
        else:
            prev = current
            current = current.next        # only advance if no deletion
 
    return head
 
```

```python
# Print Reverse Linked List

def printReverse(head):
    if head is None:
        return None
    printReverse(head.next)
    print(head.data,end=" ")
    
```
